```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('predictive_maintenance.csv')
```

```python
#
# .............////Data Read////..................
#
```

```python
#Checking of any nan or missing values 
data.info()
```

```python
#Distribution of Machine Types in Dataset
# L = low quality, M = Medium quality, H = High quality
machine_type = data[['Product ID', 'Type']].groupby('Type').count().reset_index()
color = sns.color_palette('YlGnBu')
g = plt.pie(x = machine_type['Product ID'], labels= machine_type['Type'], colors = color, autopct='%.1f%%')
plt.title('Machine Type Percentage')
plt.show()
```

```python
# Distribution of failure and non failure in the dataset 
target = data[['Product ID', 'Target']].groupby('Target',as_index = False).count().rename(columns= {'Product ID': 'Count'})
display(target)
```

```python
#Frequency of failure types
fail_type = data[['Failure Type', 'Target']].groupby('Failure Type', as_index = False).count().sort_values('Target', ascending=False).reset_index().rename(columns = {'Target': 'Count'})
display(fail_type)
```

```python
fail_machine = data[['Failure Type', 'Target']].loc[(data['Failure Type'] != 'No Failure')].groupby('Failure Type').count().reset_index()
color = sns.color_palette('YlGnBu')
g = plt.pie(x = fail_machine['Target'], labels= fail_machine['Failure Type'], colors = color, autopct='%.1f%%')
plt.title('Percentage of Machine Failure Type')
plt.show()
```

```python
print('"No Failure" Type but Failure Occured')
nofailure_fail = data.query("Target == 1 and `Failure Type` == 'No Failure'")[['Failure Type', 'Target']] 
display(nofailure_fail)
```

```python
print('Failure Occured but not Registered')
failure_nofail = data.query("Target == 0 and `Failure Type` != 'No Failure'")[['Failure Type', 'Target']] 
display(failure_nofail)
```

```python
sns.relplot(data = data, x = 'Torque [Nm]', y= 'Rotational speed [rpm]', hue = 'Failure Type', col = 'Type', palette = 'YlGnBu')
plt.suptitle('Relationship Between Torque and Rotational Speed by Machine Type', y= 1.2)
sns.despine()
plt.show()
```

```python
columns_number = data.drop(columns=['Target', 'UDI']).select_dtypes(exclude=['object']).columns

plt.figure(figsize=(25, 50))
sns.set_style("white")
col_count = 1

for col in columns_number[:10]:
    plt.subplot(8, 2, col_count)
    sns.histplot(x=data[col], kde=False, bins=10)
    
    plt.title(f'Histogram {col}')
    plt.xlabel('')
    plt.ylabel('')

    col_count += 1

sns.despine()
```
